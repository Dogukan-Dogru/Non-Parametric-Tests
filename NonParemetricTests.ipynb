# Mann-Whitney U Test

## Hypothesis


There is a significant difference in the "Credit_Limit" between customers who have churned ("Attrition_Flag" equals "Attrited Customer") and those who have not ("Attrition_Flag" equals "Existing Customer").

```python
# Filter data for existing and attrited customers
existing_customers = df[df['Attrition_Flag'] == 'Existing Customer']
attrited_customers = df[df['Attrition_Flag'] == 'Attrited Customer']
print(existing_customers)
print(attrited_customers)
```

```python
# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(existing_customers['Credit_Limit'], attrited_customers['Credit_Limit'])
```

```python
# Print the results
print(f"Mann-Whitney U Statistic: {statistic}")
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The null hypothesis is rejected. There is a significant difference in Credit Limit between existing and attrited customers.")
else:
    print("The null hypothesis is not rejected. There is no significant difference in Credit Limit between existing and attrited customers.")
```

# Chi-Square Test

## Hypothesis


There's a significant association between the "Attrition_Flag" (churn status) and another categorical variable, such as "Gender."

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Attrition_Flag'], df['Gender'])
print(contingency_table)
```

```python
# Perform the chi-square test
chi2, p_value, _, _ = chi2_contingency(contingency_table)
```

```python
# Print the results
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The null hypothesis is rejected. There is a significant association between Attrition_Flag and Gender.")
else:
    print("The null hypothesis is not rejected. There is no significant association between Attrition_Flag and Gender.")
```

# Wilcoxon Signed-Rank Test

## Hypothesis

Compare the "Credit_Limit" of customers before and after a certain event.  

```python
#Select a random subset of customers
subset_customers = df.sample(n=100, random_state=42)
```

```python
# Create a new column for the hypothetical "Credit_Limit_After_Event"
subset_customers['Credit_Limit_After_Event'] = subset_customers['Credit_Limit'] * 0.9  # Assuming a 10% decrease
print(subset_customers['Credit_Limit_After_Event'])
```

```python
# Perform the Wilcoxon signed-rank test
statistic, p_value = wilcoxon(subset_customers['Credit_Limit'], subset_customers['Credit_Limit_After_Event'])
```

```python
# Print the results
print(f"Wilcoxon signed-rank statistic: {statistic}")
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The null hypothesis is rejected. There is a significant difference in Credit_Limit before and after the event.")
else:
    print("The null hypothesis is not rejected. There is no significant difference in Credit_Limit before and after the event.")
```

Investigating the relationship between "Customer_Age" and "Total_Trans_Amt"

# Kruskal-Wallis Test

## Hypothesis

There are significant differences in the "Total_Trans_Amt" among different income categories ("Income_Category").


```python
# Subset of the data
subset_data = df[['Income_Category', 'Total_Trans_Amt']]
#print(subset_data)
gk = subset_data.groupby('Income_Category')
gk.head()
```

```python
# Perform the Kruskal-Wallis test
statistic, p_value = kruskal(
    subset_data.loc[subset_data['Income_Category'] == 'Unknown', 'Total_Trans_Amt'],
    subset_data.loc[subset_data['Income_Category'] == 'Less than $40K', 'Total_Trans_Amt'],
    subset_data.loc[subset_data['Income_Category'] == '$40K - $60K', 'Total_Trans_Amt'],
    subset_data.loc[subset_data['Income_Category'] == '$60K - $80K', 'Total_Trans_Amt'],
    subset_data.loc[subset_data['Income_Category'] == '$80K - $120K', 'Total_Trans_Amt'],
    subset_data.loc[subset_data['Income_Category'] == '$120K +', 'Total_Trans_Amt']
)
```

```python
# Print the results
print(f"Kruskal-Wallis statistic: {statistic}")
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The null hypothesis is rejected. There are significant differences in Total_Trans_Amt among income categories.")
else:
    print("The null hypothesis is not rejected. There are no significant differences in Total_Trans_Amt among income categories.")
```

# Friedman Test

## Hypothesis

There are significant differences in "Credit_Limit" across different income categories ("Income_Category").


```python
# Subset of the data for this example
subset_data = df[['Income_Category', 'Credit_Limit']]
print(subset_data)
```

```python
# Define a common sample size
sample_size = min(subset_data['Income_Category'].value_counts())
print(sample_size)
```

```python
# Create samples with equal size for each income category
samples = [subset_data.loc[subset_data['Income_Category'] == category, 'Credit_Limit'].sample(sample_size, random_state=42) for category in subset_data['Income_Category'].unique()]
```

```python
# Perform the Friedman test
statistic, p_value = friedmanchisquare(*samples)

# Print the results
print(f"Friedman test statistic: {statistic}")
print(f"P-value: {p_value}")
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("The null hypothesis is rejected. There are significant differences in Credit_Limit across different income categories.")
else:
    print("The null hypothesis is not rejected. There are no significant differences in Credit_Limit across different income categories.")
```
